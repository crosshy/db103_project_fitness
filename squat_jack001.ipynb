```python
import os, json, math, time
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#open all json files and  make DataFrame
joint_filted_y = []
sq_path ='/home/jackho/db103_project_diamond/squat_data/jack02/'
df = pd.DataFrame( columns = ['frame_count' , 'joint_num', 'move_x' , 'move_y'])
file_list = sorted(os.listdir(sq_path))
for frame_count, frame_json in enumerate(file_list):
    with open( sq_path + frame_json)as opened_file:
        if str(opened_file)[-4:] == 'json':
            joint_info = json.load(opened_file)['part_candidates'][0]
#         else:
        elif str(opened_file)[-3:] == 'png':
            continue
            for joint_num in range(15):
                jt_loop = joint_info[str(joint_num)]
                if  len(jt_loop) > 3:
                    if jt_loop[2] <  jt_loop[5]:
                        joint_filted = list(np.around(jt_loop[3:5],2))
                    else:
                        joint_filted = list(np.around(jt_loop[0:2],2))
                elif len(jt_loop) < 3:
                    pass
                else:
                     joint_filted = list(np.around(jt_loop[0:2],2))
                s = pd.Series([frame_count,  joint_num, joint_filted[0], joint_filted[1]], 
                            index  = ['frame_count' , 'joint_num', 'move_x' , 'move_y'])
                df = df.append(s, ignore_index = True)
        df.sort_values(['joint_num', 'frame_count'], ascending=[True, True])
```

```python
#get certain joint Data
def select_joint(i):
    point_data =df ['joint_num'] == i
    return df[point_data]
# select_joint(0)
```

```python
#get all differences of joints y position
def joint_dif_y(i):
    point_data =df ['joint_num'] == i
    list_y =  df[point_data]['move_y'].values.tolist()
    list_y_dif = list(np.around( [abs(j-k) for k, j in zip(list_y,list_y[1:])]))
    return list_y_dif
# print(joint_dif_y(1))
```
